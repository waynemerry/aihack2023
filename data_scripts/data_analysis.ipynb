```python
import pandas as pd

excel_data = pd.read_excel('..\data\Data for model.xlsx', index_col=0)  

gk = excel_data.groupby('user_id')
groups = dict(list(gk))

def get_time(time_string):

    if time_string == "MORNING":
        return pd.Timedelta(hours = 10)
    elif time_string == "EVENING":
        return pd.Timedelta(hours = 18)
    else:
        print("Invalid Time string")
        return pd.Timedelta(hours = 0)
    
class Analysis:

    def __init__(self, name, intial_pain_scale, initial_mood_scale, recorded_time, date_of_birth, time_string):

        self.intial_pain_scale  = intial_pain_scale
        self.initial_mood_scale = initial_mood_scale

        self.pain_scale_worsen_from_last  = []
        self.mood_scale_worsen_from_last = []
    
        self.pain_scale_worsen_from_min  = []
        self.mood_scale_worsen_from_max = []
        
        self.min_pain_scale = self.intial_pain_scale
        self.max_mood_scale = self.initial_mood_scale
        
        self.pain_scale_worsen_diff  = []
        self.mood_scale_worsen_diff = []

        self.previous_pain_scale = self.intial_pain_scale
        self.previous_mood_scale = self.initial_mood_scale

        self.pain_scale_list = [intial_pain_scale]
        self.mood_scale_list = [initial_mood_scale]

        self.time_list = [recorded_time + get_time(time_string)]
        self.age = pd.Timestamp('now').year - date_of_birth.year

        self.time_diff_list  = []
        self.name = name


    def update(self, pain_scale, mood_scale, recorded_time, time_string):

        self.pain_scale_worsen_from_last.append(pain_scale >  self.previous_pain_scale)
        self.mood_scale_worsen_from_last.append(mood_scale <  self.previous_mood_scale)

        self.pain_scale_worsen_diff.append(self.previous_pain_scale - pain_scale)
        self.mood_scale_worsen_diff.append(mood_scale  - self.previous_mood_scale)
        
        self.pain_scale_worsen_from_min.append(pain_scale >  self.min_pain_scale)
        self.mood_scale_worsen_from_max.append( mood_scale <  self.max_mood_scale)

        self.min_pain_scale = min(self.min_pain_scale, pain_scale)
        self.max_mood_scale = max(self.max_mood_scale, mood_scale)
        
        self.previous_pain_scale = pain_scale
        self.previous_mood_scale = mood_scale

        self.pain_scale_list.append(pain_scale)
        self.mood_scale_list.append(mood_scale)
        
        current_time = recorded_time + get_time(time_string)

        self.time_diff_list.append(current_time - self.time_diff_list[-1])

        self.time_list.append(current_time)
    
    def get_status(self, recorded_time):

        status  = f"Patient:{self.name}, Age:{self.age}\n\n"



def analyse_case(name, df):

    time_threshold = 2

    anlys = Analysis(name, df["pain_scale"][0], df["mood_scale"][0], df['date'][0], df['DOB'][0],  df['time_of_day'][0])
    
    for i in range(1, len(df)):
        
        anlys.update(df["pain_scale"][i], df["mood_scale"][i], df['date'][0],  df['time_of_day'][0])
        

for key, df in groups.items():
    if len(df) >= 3:
        analyse_case(key, df)    
        
```
